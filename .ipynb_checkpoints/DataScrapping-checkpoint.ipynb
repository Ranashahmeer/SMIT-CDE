```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://realpython.github.io/fake-jobs/'
request = requests.get(url)
# request.content
```

```python
soup=BeautifulSoup(request.content,'html.parser')
```

```python
container = soup.find('ResultsContainer')
```

```python
job = soup.find(class_='title is-5')
job.text
```

```python
subtitle = soup.find(class_='subtitle is-6 company')
subtitle.text
```

```python
loc = soup.find(class_='location')
loc.text.strip()
```

```python
date = soup.find(class_='is-small has-text-grey')
date.text
```

```python
secJob = soup.find(class_='title is-5')
secJob.text
```

```python
allJobs= soup.find_all(class_='title is-5')
allJobs[0].text
allJobs[1].text
```

```python
Container =soup.find(id = 'ResultsContainer')
jobDivs= Container.find_all(class_ = 'column is-half')
```

```python
i=0
for job in jobDivs:
    i+=1
    print(f'{i}) Job')
    title=(job.find(class_='title is-5')).text
    subtitle = (job.find(class_='subtitle is-6 company')).text
    loc = (soup.find(class_='location')).text.strip()
    date = (soup.find(class_='is-small has-text-grey')).text.strip()
    
    print(f"Title : {title},\nSubtitle : {subtitle}\nLocation {loc},\nDate : {date}")
    break
```

```python
url='https://www.pakwheels.com/new-cars/audi/'
# for url in url:
req = requests.get(url)
soup=BeautifulSoup(req.content,'html.parser')
soup
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
container =soup.find(class_ ='generic-car-widgets-container')
carList=container.find_all(class_ ='col-md-3')
len(carList)
```

```python
url=[]
container =soup.find_all(class_ ='generic-car-widgets-container')
for container in container:
    carList=container.find_all(class_ ='col-md-3')
    for carURL in carList:
        tag=carURL.find('a')
        link = 'https://www.pakwheels.com' + tag.get('href')
        url.append(link)
len(url)
```
