Case Study Data Visualisation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('gapminder.csv')
```

```python
life_exp = df.life_exp
gdp_cap = df.gdp_cap
pop=df['population'].to_numpy()

plt.scatter(gdp_cap,life_exp, s=(pop / 1e6)*3, alpha=0.8 )
plt.xscale('log')

plt.title('World Development Graph')
plt.xlabel('GDP per Capita [In USD]')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.xticks([1000,10000,100000],['1k','10k','100k'])
```
