#Matplot lib

```python
import matplotlib.pyplot as plt
```

```python
year = [1995,1996,1997,1998,1999]
pop =[2.24,2.2,2.68,2.29,2.27]
plt.plot(year,pop)
plt.xlabel('year')
plt.ylabel('pop')
plt.title('year/pop')
plt.yticks(0,2,4,6,8)
plt.show()
```

```python
plt.scatter(year,pop)
plt.xlabel('year')
plt.ylabel('pop')
plt.title('year/pop')
```
